# Q3) Transformer Failures

## 1. Construct a probability space \(($\Omega$, $\mathcal{F}$, P)\)

We model the number of transformer failures in a day.

### Sample Space
\[
$\Omega$ = \{0,1,2,3\}
\]

where:
- \(0\) = no failures
- \(1\) = one failure
- \(2\) = two failures
- \(3\) = three failures

### Sigma-algebra
Since \($\Omega$\) is finite, we can take:
\[
$\mathcal{F}$ = 2^$\Omega$
\]
(the power set of \($\Omega$\)).

### Probability Measure
The probabilities are:
\[
P(0)=0.50,$\quad$ P(1)=0.30,$\quad$ P(2)=0.15,$\quad$ P(3)=0.05
\]

Check:
\[
0.50+0.30+0.15+0.05=1
\]

So this is a valid probability space.

---

## 2. Define \(X\) as a random variable

Define the random variable:
\[
X:$\Omega$ $\to$ $\mathbb{R}$
\]

by
\[
X($\omega$)=$\omega$
\]

Thus:
\[
X(0)=0,$\quad$ X(1)=1,$\quad$ X(2)=2,$\quad$ X(3)=3
\]

---

## 3. Probability Mass Function (PMF)

The probability mass function is:

\[
p_X(x)=P(X=x)
\]

So,

\[
p_X(x)=
\begin{cases}
0.50 & x=0 \\
0.30 & x=1 \\
0.15 & x=2 \\
0.05 & x=3 \\
0 & \text{otherwise}
\end{cases}
\]

---

## 4. Compute the Expectation \(E[X]\)

The expectation of a discrete random variable is:

E[X] = Σ xP(X=x)

Substitute the values:

\[
E[X]
=0(0.50)+1(0.30)+2(0.15)+3(0.05)
\]

\[
=0+0.30+0.30+0.15
\]

\[
E[X]=0.75
\]

### Answer:
\
$\boxed{E[X]=0.75}$


---

## 5. Compute the Variance \(Var(X)\)

First compute E[X²]:

E[X²] = 0²(0.50) + 1²(0.30) + 2²(0.15) + 3²(0.05)

= 0 + 0.30 + 0.60 + 0.45

E[X²] = 1.35

Variance formula:

Var(X) = E[X²] − (E[X])²

Var(X) = 1.35 − (0.75)²

= 1.35 − 0.5625

\[
Var(X)=0.7875
\]

### Answer:

$\boxed{Var(X)=0.7875}$

- The variance \(Var(X)=0.7875\) measures how much the number of failures varies from day to day. A larger variance indicates more unpredictability in transformer failures.

# Q4) Probability of Defective Items

1. Construct a probability space $(\Omega, \mathcal{F}, P)$ suitable for this model.To model the proportion of defective items $X$, we define the probability space $(\Omega, \mathcal{F}, P)$ as follows:

Sample Space ($\Omega$): The set of all possible outcomes for the proportion of defective items. Since the proportion must lie between $0$ and $1$:$$\Omega = [0, 1]$$$\sigma$-algebra ($\mathcal{F}$): The collection of measurable events. For a continuous interval on the real line, we use the Borel $\sigma$-algebra restricted to $[0, 1]$:$$\mathcal{F} = \mathcal{B}([0, 1])$$Probability Measure ($P$): For any event $A \in \mathcal{F}$, the probability is defined by integrating the given probability density function $f(x)$ over $A$:






$$P(A) = \int_{A} f(x) \, dx = \int_{A \cap [0,1]} \frac{1}{B(2,18)} x^{1} (1-x)^{17} \, dx$$









---
2. Define $X$ as a continuous random variable on this space.

The random variable $X$ is defined as the identity function mapping the sample space to the real numbers:

$$X: \Omega \to \mathbb{R} \quad \text{defined by} \quad X(\omega) = \omega \quad \text{for all } \omega \in \Omega$$

Since $\Omega = [0,1]$, $X$ is a continuous random variable taking values in the interval $[0, 1]$, representing the proportion of defective items in a batch.


---

3. Write down the probability density function of $X$.

The probability density function (PDF) of $X \sim \text{Beta}(\alpha = 2, \beta = 18)$ is:


$$f(x) = \begin{cases}
\frac{1}{B(2, 18)} x (1 - x)^{17}, & 0 < x < 1, \
0, & \text{otherwise}
\end{cases}$$

where the normalization constant is the Beta function:

$$B(2, 18) = \frac{\Gamma(2)\Gamma(18)}{\Gamma(20)} = \frac{1! \cdot 17!}{19!} = \frac{1}{19 \times 18} = \frac{1}{342}$$



Thus, the simplified PDF for $0 < x < 1$ is:$$f(x) = 342x(1-x)^{17}$$

---


4. Verify that $f$ is a valid probability density function.For $f(x)$ to be a valid PDF, it must satisfy two conditions:

    1) $f(x) \geq 0$ for all $x \in \mathbb{R}$.

    2) $\int_{-\infty}^{\infty} f(x) \, dx = 1$.

Condition 1: Non-negativity

For $0 < x < 1$, both $x > 0$ and $(1-x) > 0$. Since $B(2,18) > 0$, we have:


$$f(x) = 342x(1-x)^{17} \geq 0$$


For $x$ outside $(0,1)$, $f(x) = 0 \geq 0$. Thus, $f(x) \geq 0$ everywhere.



Condition 2: Integration to 1

Using the definition of the Beta function $\int_{0}^{1} x^{\alpha-1}(1-x)^{\beta-1} \, dx = B(\alpha, \beta)$:



$$\int_{-\infty}^{\infty} f(x) \, dx = \int_{0}^{1} \frac{1}{B(2, 18)} x^{2-1} (1-x)^{18-1} \, dx = \frac{1}{B(2, 18)} \int_{0}^{1} x^{1} (1-x)^{17} \, dx$$




$$\int_{-\infty}^{\infty} f(x) \, dx = \frac{B(2, 18)}{B(2, 18)} = 1$$


Both conditions are verified.

---


5. Compute the expectation $E[X]$.

For a Beta-distributed random variable $X \sim \text{Beta}(\alpha, \beta)$, the expectation is given by:




$$E[X] = \frac{\alpha}{\alpha + \beta}$$


Substituting $\alpha = 2$ and $\beta = 18$:


$$E[X] = \frac{2}{2 + 18} = \frac{2}{20} = 0.10$$


---


6. Compute the variance $\text{Var}(X)$.

The variance of a Beta distribution is given by:

$$\text{Var}(X) = \frac{\alpha\beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}$$


Substituting $\alpha = 2$ and $\beta = 18$:


$$\text{Var}(X) = \frac{2 \times 18}{(2 + 18)^2 (2 + 18 + 1)} = \frac{36}{(20)^2 (21)} = \frac{36}{400 \times 21} = \frac{36}{8400} = \frac{3}{700} \approx 0.004286$$


---


7. Find the probability $P(X \leq 0.10)$.


We need to compute:


$$P(X \leq 0.10) = \int_{0}^{0.10} f(x) \, dx = \frac{1}{B(2, 18)} \int_{0}^{0.10} x(1-x)^{17} \, dx$$



In Beta integral (Incomplete Beta Function) form:



$$P(X \leq 0.10) = \frac{B(0.10; 2, 18)}{B(2, 18)} = I_{0.10}(2, 18)$$



Exact integration evaluation:


Using $f(x) = 342x(1-x)^{17}$ and integration by parts (letting $u = x \implies du = dx$ and $dv = (1-x)^{17}dx \implies v = -\frac{(1-x)^{18}}{18}$):



$$\int x(1-x)^{17} \, dx = -\frac{x(1-x)^{18}}{18} - \int -\frac{(1-x)^{18}}{18} \, dx = -\frac{x(1-x)^{18}}{18} - \frac{(1-x)^{19}}{18 \times 19}$$



Multiplying by the constant $342 = 18 \times 19$:


$$\int 342x(1-x)^{17} \, dx = -19x(1-x)^{18} - (1-x)^{19}$$


Evaluating from $0$ to $0.10$:



$$P(X \leq 0.10) = \left[ -19(0.10)(0.90)^{18} - (0.90)^{19} \right] - \left[ 0 - 1 \right]$$


$$P(X \leq 0.10) = 1 - 1.9(0.90)^{18} - (0.90)^{19}$$



$$P(X \leq 0.10) = 1 - (0.90)^{18} \cdot [1.9 + 0.90] = 1 - 2.8(0.90)^{18}$$


Using $(0.90)^{18} \approx 0.15009$:


$$P(X \leq 0.10) \approx 1 - 2.8(0.15009) = 1 - 0.42025 = 0.57975$$


---



8. Find the probability $P(0.05 \leq X \leq 0.15)$.


We need to compute:


$$P(0.05 \leq X \leq 0.15) = \int_{0.05}^{0.15} f(x) \, dx = \frac{1}{B(2, 18)} \int_{0.05}^{0.15} x(1-x)^{17} \, dx$$



In Beta integral form:


$$P(0.05 \leq X \leq 0.15) = I_{0.15}(2, 18) - I_{0.05}(2, 18)$$


Exact integration evaluation:


Using the antiderivative $F(x) = -19x(1-x)^{18} - (1-x)^{19}$:



$$P(0.05 \leq X \leq 0.15) = F(0.15) - F(0.05)$$



$$F(0.15) = -19(0.15)(0.85)^{18} - (0.85)^{19} = -2.85(0.85)^{18} - 0.85(0.85)^{18} = -3.7(0.85)^{18} \approx -3.7(0.0536) \approx -0.1983$$



$$F(0.05) = -19(0.05)(0.95)^{18} - (0.95)^{19} = -0.95(0.95)^{18} - 0.95(0.95)^{18} = -1.9(0.95)^{18} \approx -1.9(0.3972) \approx -0.7547$$



$$P(0.05 \leq X \leq 0.15) \approx -0.1983 - (-0.7547) = 0.5564$$

# Q5) Expected Project Completion Time: Solutions



1. Specify the support of $X$.


The support of a random variable is the set of all values for which the probability density function is non-zero. For a Triangular distribution with parameters $a = 4$ and $b = 10$:


$$\text{Support of } X = [4, 10]$$



This represents the interval of possible hours needed to complete the installation.



---



2. Write down the probability density function $f_X(x)$.



For a general Triangular distribution with lower limit $a$, upper limit $b$, and mode $c$ ($a < c < b$), the probability density function is:




$$f_X(x) = \begin{cases}
\frac{2(x - a)}{(b - a)(c - a)}, & a \leq x \leq c \
\frac{2(b - x)}{(b - a)(b - c)}, & c < x \leq b \
0, & \text{otherwise}
\end{cases}$$





Substituting our parameters $a = 4$, $b = 10$, and $c = 6$:



For $4 \leq x \leq 6$:




$$\frac{2(x - 4)}{(10 - 4)(6 - 4)} = \frac{2(x - 4)}{6 \times 2} = \frac{x - 4}{6}$$




For $6 < x \leq 10$:




$$\frac{2(10 - x)}{(10 - 4)(10 - 6)} = \frac{2(10 - x)}{6 \times 4} = \frac{10 - x}{12}$$




Thus, the probability density function is:




$$f_X(x) = \begin{cases}
\frac{x - 4}{6}, & 4 \leq x \leq 6 \
\frac{10 - x}{12}, & 6 < x \leq 10 \
0, & \text{otherwise}
\end{cases}$$



---



3. Verify that $f_X$ integrates to $1$ over its support.




To verify, we integrate $f_X(x)$ over the support $[4, 10]$ by breaking it into two intervals:





$$\int_{4}^{10} f_X(x) \, dx = \int_{4}^{6} \frac{x - 4}{6} \, dx + \int_{6}^{10} \frac{10 - x}{12} \, dx$$





Part 1: Integrating over $[4, 6]$




$$\int_{4}^{6} \frac{x - 4}{6} \, dx = \frac{1}{6} \left[ \frac{(x - 4)^2}{2} \right]_{4}^{6} = \frac{1}{12} \left[ (6 - 4)^2 - (4 - 4)^2 \right] = \frac{4}{12} = \frac{1}{3}$$





Part 2: Integrating over $[6, 10]$





$$\int_{6}^{10} \frac{10 - x}{12} \, dx = \frac{1}{12} \left[ -\frac{(10 - x)^2}{2} \right]_{6}^{10} = -\frac{1}{24} \left[ (10 - 10)^2 - (10 - 6)^2 \right] = -\frac{1}{24} [0 - 16] = \frac{16}{24} = \frac{2}{3}$$





Total Integral



$$\int_{4}^{10} f_X(x) \, dx = \frac{1}{3} + \frac{2}{3} = 1$$




The density function integrates to $1$, verifying it is a valid PDF.




---







4. Compute the expectation $E[X]$.




The expectation (mean) of a Triangular distribution is given by:



$$E[X] = \frac{a + b + c}{3}$$





Substituting $a = 4$, $b = 10$, and $c = 6$:





$$E[X] = \frac{4 + 10 + 6}{3} = \frac{20}{3} \approx 6.67 \text{ hours}$$




---




5. Compute the variance $\text{Var}(X)$.




The variance of a Triangular distribution is given by:






$$\text{Var}(X) = \frac{a^2 + b^2 + c^2 - ab - ac - bc}{18}$$




Substituting $a = 4$, $b = 10$, and $c = 6$:





$$\text{Var}(X) = \frac{4^2 + 10^2 + 6^2 - (4 \times 10) - (4 \times 6) - (10 \times 6)}{18}$$




$$\text{Var}(X) = \frac{16 + 100 + 36 - 40 - 24 - 60}{18}$$




$$\text{Var}(X) = \frac{152 - 124}{18} = \frac{28}{18} = \frac{14}{9} \approx 1.56 \text{ hours}^2$$




---



6. Interpret the values of $a$, $b$, and $c$ in the context of this problem.



$a = 4$ hours (Optimistic Time): This is the minimum possible time required to complete the installation. Under perfect, ideal conditions with no delays or complications, the team can complete the task in $4$ hours.



$b = 10$ hours (Pessimistic Time): This is the maximum possible time required to complete the installation. Even if major unexpected obstacles or delays occur, the task will take no longer than $10$ hours.




$c = 6$ hours (Most Likely Time): This is the mode of the distribution. It represents the typical duration of this task under normal working conditions.

7. Explain briefly why a Triangular distribution is a reasonable model in this setting.



A Triangular distribution is a highly practical model for project management scenarios (such as PERT/CPM analysis) because:




Intuitive Parameters: Project planners and domain experts can easily estimate the best-case ($a$), worst-case ($b$), and most likely ($c$) scenarios, even when historical statistical data is scarce.




Limited Support: Unlike a Normal distribution, which extends infinitely, a Triangular distribution has finite limits. This prevents unrealistic modeling outcomes (like negative times or infinitely long project durations).

8. Compute the probability that the installation takes at most $5$ hours, that is, compute $P(X \leq 5)$.






Since $5$ lies in the interval $[4, 6]$, we integrate the first segment of the PDF from $4$ to $5$:




$$P(X \leq 5) = \int_{4}^{5} \frac{x - 4}{6} \, dx$$





Using the antiderivative evaluated earlier:





$$P(X \leq 5) = \left[ \frac{(x - 4)^2}{12} \right]_{4}^{5} = \frac{(5 - 4)^2}{12} - \frac{(4 - 4)^2}{12} = \frac{1}{12} \approx 0.0833 \text{ (or } 8.33\% \text{)}$$

9. Compute the probability that the installation takes between $5$ and $8$ hours, that is, compute $P(5 \leq X \leq 8)$.




We split the probability calculation across the transition boundary $c = 6$:






$$P(5 \leq X \leq 8) = P(5 \leq X \leq 6) + P(6 < X \leq 8)$$






Part 1: $P(5 \leq X \leq 6)$




$$P(5 \leq X \leq 6) = \int_{5}^{6} \frac{x - 4}{6} \, dx = \left[ \frac{(x - 4)^2}{12} \right]_{5}^{6} = \frac{(6-4)^2}{12} - \frac{(5-4)^2}{12} = \frac{4}{12} - \frac{1}{12} = \frac{3}{12} = \frac{1}{4}$$



Part 2: $P(6 < X \leq 8)$




Using the second piece of our PDF $\frac{10-x}{12}$ on $(6, 10]$:




$$P(6 < X \leq 8) = \int_{6}^{8} \frac{10 - x}{12} \, dx = \left[ -\frac{(10 - x)^2}{24} \right]_{6}^{8} = -\frac{(10 - 8)^2}{24} - \left( -\frac{(10 - 6)^2}{24} \right)$$




$$P(6 < X \leq 8) = -\frac{4}{24} + \frac{16}{24} = \frac{12}{24} = \frac{1}{2}$$




Total Probability$$P(5 \leq X \leq 8) = \frac{1}{4} + \frac{1}{2} = \frac{3}{4} = 0.75 \text{ (or } 75\% \text{)}$$

$$6c = a + 4c + b$$

 Analytical behavior of the variance



 If the expected value matches the mode ($E[X] = c$), the variance reaches its maximum possible value for any fixed interval width $(b - a)$.


 In any Beta-based distribution (like PERT) over a fixed interval $[a, b]$, shifting the mode $c$ away from the center (making the distribution skewed) pulls the bulk of the probability mass closer to one of the boundaries. This tightens the distribution on that side, which decreases the overall variance.



 When $E[X] = c$ (making the distribution perfectly symmetric), the probability density is spread out as evenly as possible around the center without skewness, which maximizes the uncertainty (variance) for that specific span.
