# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius $R$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

---

## 1. Uniform distribution with respect to area

If the point is uniformly distributed over the area of the disk, then probabilities are proportional to areas.

The event $A$ corresponds to the outer annulus between radii $R-1$ and $R$.

The area of the whole disk is

$$
\pi R^2
$$

The area of the inner disk of radius $R-1$ is

$$
\pi (R-1)^2
$$

Hence, the area of the annulus is

$$
\pi R^2 - \pi (R-1)^2
$$

Therefore,

$$
P(A)
=
\frac{\pi R^2 - \pi (R-1)^2}{\pi R^2}
$$

Simplifying,

$$
P(A)
=
\frac{R^2 - (R-1)^2}{R^2}
$$

Since

$$
(R-1)^2 = R^2 - 2R + 1
$$

we get

$$
P(A)
=
\frac{R^2 - (R^2 - 2R + 1)}{R^2}
$$

$$
P(A)
=
\frac{2R - 1}{R^2}
$$

Thus,

$$
\boxed{
P(A)=\frac{2R-1}{R^2}
}
$$

---

## 2. Radius chosen uniformly from $[0,R]$

Now assume the radius $r$ is chosen uniformly from $[0,R]$ and the angle $\theta$ is chosen independently and uniformly from $[0,2\pi]$.

The event $A$ occurs when

$$
R-1 \le r \le R
$$

Since $r$ is uniformly distributed on $[0,R]$,

$$
P(A)
=
\frac{\text{length of }[R-1,R]}{\text{length of }[0,R]}
$$

$$
P(A)
=
\frac{1}{R}
$$

Hence,

$$
\boxed{
P(A)=\frac{1}{R}
}
$$

---

## 3. Why are the answers different?

The two probability models are different.

In the first model, the point is uniformly distributed with respect to area. Since outer rings have larger area than inner rings, points are more likely to lie near the boundary.

In polar coordinates, the area element is

$$
dA = r\,dr\,d\theta
$$

which shows that larger values of $r$ correspond to larger area.

In the second model, the radius is chosen uniformly, meaning every radius value is equally likely. This does not produce a uniform distribution over area because thin rings near the center and near the boundary are treated equally despite having different areas.

Therefore, the two models produce different probabilities.

# Q2) Bertrand's Paradox

A chord of a circle is chosen “at random.” We wish to find the probability that its length exceeds the side length of the inscribed equilateral triangle.

Let the circle have radius $R$.

The side length of an inscribed equilateral triangle is

$$
\sqrt{3}R
$$

A chord is longer than this exactly when its perpendicular distance from the center is less than

$$
\frac{R}{2}
$$

This follows from the chord-length formula

$$
L = 2\sqrt{R^2-d^2}
$$

where $d$ is the perpendicular distance from the center to the chord.

We require

$$
2\sqrt{R^2-d^2} > \sqrt{3}R
$$

Squaring both sides,

$$
4(R^2-d^2) > 3R^2
$$

$$
R^2 > 4d^2
$$

$$
d < \frac{R}{2}
$$

Thus, the problem reduces to finding the probability that the distance of the chord from the center is less than $R/2$.

---

# Model 1: Random endpoints on the circumference

Choose two independent random points on the circumference and join them.

Fix one endpoint and let $\theta$ be the angle subtended at the center.

The chord exceeds the triangle side length when

$$
\theta > \frac{2\pi}{3}
$$

Since $\theta$ is uniformly distributed on $[0,\pi]$,

the favorable interval has length

$$
\pi - \frac{2\pi}{3}
=
\frac{\pi}{3}
$$

Hence,

$$
P
=
\frac{\pi/3}{\pi}
=
\frac{1}{3}
$$

Therefore,

$$
\boxed{
P=\frac{1}{3}
}
$$

---

# Model 2: Random radius and random midpoint on the radius

Choose a radius at random and then choose the midpoint of the chord uniformly along that radius.

The chord is longer than the triangle side precisely when the midpoint lies within distance

$$
\frac{R}{2}
$$

from the center.

Since the midpoint is uniformly distributed along the radius interval $[0,R]$,

$$
P
=
\frac{R/2}{R}
=
\frac{1}{2}
$$

Therefore,

$$
\boxed{
P=\frac{1}{2}
}
$$

---

# Model 3: Midpoint uniformly distributed over the disk

Choose the midpoint of the chord uniformly over the entire disk.

The chord is longer than the triangle side exactly when the midpoint lies inside the concentric disk of radius

$$
\frac{R}{2}
$$

Since probability is proportional to area,

$$
P
=
\frac{\pi (R/2)^2}{\pi R^2}
$$

$$
P
=
\frac{R^2/4}{R^2}
=
\frac{1}{4}
$$

Therefore,

$$
\boxed{
P=\frac{1}{4}
}
$$

---

# Conclusion

Different mathematically natural definitions of “random chord” produce different probability distributions on the set of chords.

Thus Bertrand’s paradox shows that the phrase

> “choose a chord at random”

is mathematically ambiguous unless the method of randomization is specified.

The three natural models give:

$$
\boxed{
\frac{1}{3},\quad \frac{1}{2},\quad \frac{1}{4}
}
$$

showing that different assumptions lead to different answers.

# Q8) Probability of Being a Girl

A family has two children. Assume each child is equally likely to be a boy or a girl, independently of the other.

We want to find the probability that both children are girls, given that at least one of them is a girl.

---

Let

$$
A = \{\text{both children are girls}\}
$$

and

$$
B = \{\text{at least one child is a girl}\}
$$

The sample space for two children is

$$
\{BB, BG, GB, GG\}
$$

where

- $BB$ = both boys
- $BG$ = older boy, younger girl
- $GB$ = older girl, younger boy
- $GG$ = both girls

Since each child is equally likely to be a boy or girl independently,

each outcome has probability

$$
\frac{1}{4}
$$

---

## Step 1: Condition on at least one girl

The event “at least one girl” excludes only the outcome $BB$.

Thus the reduced sample space is

$$
\{BG, GB, GG\}
$$

Each of these outcomes is equally likely.

---

## Step 2: Count favorable outcomes

The event “both children are girls” corresponds only to

$$
GG
$$

Thus there is 1 favorable outcome out of 3 possible outcomes.

Therefore,

$$
P(A \mid B)
=
\frac{1}{3}
$$

Hence,

$$
\boxed{
P(\text{both girls} \mid \text{at least one girl})
=
\frac{1}{3}
}
$$

# Q9) Information Available in a Discrete Random Variable

A factory inspects a product and records its quality using two different coding systems.

The sample space is

$$
\Omega = \{G_1,G_2,D_1,D_2\}
$$

with probabilities

$$
P(G_1)=0.50,\qquad
P(G_2)=0.20,\qquad
P(D_1)=0.10,\qquad
P(D_2)=0.20.
$$

Two random variables are defined:

## Random Variable \(X\)

\(X\) records only whether the item is good or defective.

$$
X(G_1)=0,\qquad
X(G_2)=0,\qquad
X(D_1)=1,\qquad
X(D_2)=1
$$

## Random Variable \(Y\)

\(Y\) records both quality and machine.

$$
Y(G_1)=1,\qquad
Y(G_2)=2,\qquad
Y(D_1)=3,\qquad
Y(D_2)=4
$$

---

# 1. Probability Space

The probability space is

$$
(\Omega,\mathcal{F},P)
$$

where

$$
\Omega=\{G_1,G_2,D_1,D_2\}
$$

and

$$
\mathcal{F}=\mathcal{P}(\Omega)
$$

(the power set of \(\Omega\)).

The probability measure is defined by

$$
P(G_1)=0.50,\quad
P(G_2)=0.20,\quad
P(D_1)=0.10,\quad
P(D_2)=0.20.
$$

---

# 2. Sigma-algebra Generated by \(X\)

Since \(X\) only distinguishes between good and defective items,

the inverse images are

$$
X^{-1}(\{0\})=\{G_1,G_2\}
$$

and

$$
X^{-1}(\{1\})=\{D_1,D_2\}.
$$

Hence,

$$
\sigma(X)
=
\{
\emptyset,
\Omega,
\{G_1,G_2\},
\{D_1,D_2\}
\}.
$$

---

# 3. Sigma-algebra Generated by \(Y\)

Since \(Y\) assigns different values to every outcome,

every singleton set can be recovered.

Thus,

$$
\sigma(Y)=\mathcal{P}(\Omega).
$$

Explicitly,

$$
\sigma(Y)
=
\{
\emptyset,
\{G_1\},
\{G_2\},
\{D_1\},
\{D_2\},
\dots,
\Omega
\}.
$$

So \(Y\) generates the full power set.

---

# 4. Show that \(\sigma(X)\subseteq\sigma(Y)\)

Since

$$
\sigma(X)
=
\{
\emptyset,
\Omega,
\{G_1,G_2\},
\{D_1,D_2\}
\}
$$

and

$$
\sigma(Y)=\mathcal{P}(\Omega),
$$

every set in \(\sigma(X)\) is also contained in \(\sigma(Y)\).

Therefore,

$$
\boxed{
\sigma(X)\subseteq\sigma(Y)
}
$$

This means that \(Y\) contains more information than \(X\).

- \(X\) only tells whether the product is good or defective.
- \(Y\) tells both the quality and which machine produced it.

---

# 5. Marginal Distribution of \(X\)

\(X=0\) corresponds to good products:

$$
P(X=0)
=
P(G_1)+P(G_2)
=
0.50+0.20
=
0.70
$$

\(X=1\) corresponds to defective products:

$$
P(X=1)
=
P(D_1)+P(D_2)
=
0.10+0.20
=
0.30
$$

Thus,

$$
\boxed{
P(X=0)=0.70,\qquad P(X=1)=0.30
}
$$

---

# 6. Marginal Distribution of \(Y\)

Since each outcome has a unique code,

$$
P(Y=1)=P(G_1)=0.50
$$

$$
P(Y=2)=P(G_2)=0.20
$$

$$
P(Y=3)=P(D_1)=0.10
$$

$$
P(Y=4)=P(D_2)=0.20
$$

Hence,

$$
\boxed{
P(Y=1)=0.50,\;
P(Y=2)=0.20,\;
P(Y=3)=0.10,\;
P(Y=4)=0.20
}
$$

---

# 7. Joint Probabilities

Because each value of \(Y\) uniquely determines the outcome,

$$
P(X=0,Y=1)=P(G_1)=0.50
$$

$$
P(X=0,Y=2)=P(G_2)=0.20
$$

$$
P(X=1,Y=3)=P(D_1)=0.10
$$

$$
P(X=1,Y=4)=P(D_2)=0.20
$$

Thus,

$$
\boxed{
P(X=0,Y=1)=0.50
}
$$

$$
\boxed{
P(X=0,Y=2)=0.20
}
$$

$$
\boxed{
P(X=1,Y=3)=0.10
}
$$

$$
\boxed{
P(X=1,Y=4)=0.20
}
$$

---

# 8. Conditional Probabilities

## \(P(X=1\mid Y=3)\)

If \(Y=3\), then the outcome must be \(D_1\), so \(X=1\) with certainty.

$$
P(X=1\mid Y=3)=1
$$

---

## \(P(X=1\mid Y=4)\)

If \(Y=4\), then the outcome must be \(D_2\), so

$$
P(X=1\mid Y=4)=1
$$

---

## \(P(Y=3\mid X=1)\)

$$
P(Y=3\mid X=1)
=
\frac{P(Y=3,X=1)}{P(X=1)}
$$

$$
=
\frac{0.10}{0.30}
=
\frac{1}{3}
$$

---

## \(P(Y=4\mid X=1)\)

$$
P(Y=4\mid X=1)
=
\frac{0.20}{0.30}
=
\frac{2}{3}
$$

Hence,

$$
\boxed{
P(X=1\mid Y=3)=1
}
$$

$$
\boxed{
P(X=1\mid Y=4)=1
}
$$

$$
\boxed{
P(Y=3\mid X=1)=\frac13
}
$$

$$
\boxed{
P(Y=4\mid X=1)=\frac23
}
$$

---

# 9. Interpretation of Conditioning on \(X\) and \(Y\)

Conditioning on \(X\):

- only tells whether the product is good or defective.

Conditioning on \(Y\):

- identifies both the quality and the machine exactly.

Thus, conditioning on \(Y\) gives more precise information than conditioning on \(X\).

---

# 10. Information in \(\sigma(X)\) and \(\sigma(Y)\)

## Information in \(\sigma(X)\)

\(\sigma(X)\) contains information only about:

- whether the product is good,
- or defective.

It cannot distinguish between Machine 1 and Machine 2.

---

## Information in \(\sigma(Y)\)

\(\sigma(Y)\) contains complete information about:

- product quality,
- and the producing machine.

It distinguishes all four outcomes individually.

Thus \(Y\) is a finer observation than \(X\).

---


```python
import matplotlib.pyplot as plt

# Marginal distributions
X_labels = ['Good (X=0)', 'Defective (X=1)']
X_probs = [0.70, 0.30]

Y_labels = ['G1', 'G2', 'D1', 'D2']
Y_probs = [0.50, 0.20, 0.10, 0.20]

# Plot for X
plt.figure(figsize=(6,4))
plt.bar(X_labels, X_probs)
plt.ylabel('Probability')
plt.title('Marginal Distribution of X')
plt.ylim(0,1)
plt.show()

# Plot for Y
plt.figure(figsize=(6,4))
plt.bar(Y_labels, Y_probs)
plt.ylabel('Probability')
plt.title('Marginal Distribution of Y')
plt.ylim(0,1)
plt.show()
```

# Q10) Information Available in Continuous Random Variables

A chemical plant monitors each production run using two different measurement systems.

For a given run:

- \(T\) = deviation of reactor temperature from target,
- \(P\) = deviation of reactor pressure from target.

Assume

$$
T\sim N(0,1),
\qquad
P\sim N(0,4),
$$

and \(T\) and \(P\) are independent.

The probability space is

$$
\Omega=\mathbb R^2,
\qquad
\mathfrak F=\mathfrak B(\mathbb R^2),
$$

with probability measure induced by the joint density

$$
f_{T,P}(t,p)
=
\frac1{4\pi}
\exp\left(
-\frac{t^2}{2}
-\frac{p^2}{8}
\right).
$$

Define

$$
X(t,p)=t
$$

and

$$
Y(t,p)=(t,p).
$$

---

# 1. Probability Space

The probability space is

$$
(\Omega,\mathfrak F,P)
$$

where

$$
\Omega=\mathbb R^2
$$

is the sample space,

$$
\mathfrak F=\mathfrak B(\mathbb R^2)
$$

is the Borel \(\sigma\)-algebra on \(\mathbb R^2\),

and \(P\) is the probability measure defined through the density

$$
f_{T,P}(t,p)
=
\frac1{4\pi}
\exp\left(
-\frac{t^2}{2}
-\frac{p^2}{8}
\right).
$$

Thus for any Borel set \(A\subseteq\mathbb R^2\),

$$
P(A)
=
\int_A
f_{T,P}(t,p)\,dt\,dp.
$$

---

# 2. Sigma-algebra Generated by \(X\)

Since

$$
X(t,p)=t,
$$

\(X\) only observes the temperature coordinate.

Thus,

$$
\sigma(X)
=
\{
X^{-1}(B):B\in\mathfrak B(\mathbb R)
\}.
$$

For any Borel set \(B\subseteq\mathbb R\),

$$
X^{-1}(B)
=
B\times\mathbb R.
$$

Hence,

$$
\boxed{
\sigma(X)
=
\{
B\times\mathbb R:B\in\mathfrak B(\mathbb R)
\}
}
$$

This sigma-algebra contains all vertical strips in \(\mathbb R^2\).

---

# 3. Sigma-algebra Generated by \(Y\)

Since

$$
Y(t,p)=(t,p),
$$

\(Y\) observes both coordinates completely.

Therefore,

$$
\sigma(Y)
=
\mathfrak B(\mathbb R^2).
$$

Hence,

$$
\boxed{
\sigma(Y)=\mathfrak B(\mathbb R^2)
}
$$

---

# 4. Show that \(\sigma(X)\subseteq\sigma(Y)\)

Every set of the form

$$
B\times\mathbb R
$$

is a Borel subset of \(\mathbb R^2\).

Thus every set in \(\sigma(X)\) belongs to \(\sigma(Y)\).

Therefore,

$$
\boxed{
\sigma(X)\subseteq\sigma(Y)
}
$$

This means:

- \(X\) records only temperature deviation,
- \(Y\) records both temperature and pressure deviations.

Hence \(Y\) contains more information than \(X\).

---

# 5. Marginal Density of \(X\)

The marginal density of \(X\) is

$$
f_X(t)
=
\int_{-\infty}^{\infty}
f_{T,P}(t,p)\,dp.
$$

Substitute the joint density:

$$
f_X(t)
=
\int_{-\infty}^{\infty}
\frac1{4\pi}
\exp\left(
-\frac{t^2}{2}
-\frac{p^2}{8}
\right)
dp.
$$

Factor out the terms independent of \(p\):

$$
f_X(t)
=
\frac1{4\pi}
e^{-t^2/2}
\int_{-\infty}^{\infty}
e^{-p^2/8}\,dp.
$$

Using

$$
\int_{-\infty}^{\infty}
e^{-p^2/8}\,dp
=
2\sqrt{2\pi},
$$

we obtain

$$
f_X(t)
=
\frac1{\sqrt{2\pi}}
e^{-t^2/2}.
$$

Hence,

$$
\boxed{
f_X(t)
=
\frac1{\sqrt{2\pi}}
e^{-t^2/2}
}
$$

which is the density of \(N(0,1)\).

---

# 6. Marginal Density of \(Y\)

Since \(Y=(T,P)\),

its density is exactly the joint density:

$$
\boxed{
f_Y(t,p)
=
\frac1{4\pi}
\exp\left(
-\frac{t^2}{2}
-\frac{p^2}{8}
\right)
}
$$

---

# 7. Compute Probabilities

## (a) \(P(X\le0)\)

Since \(X\sim N(0,1)\) is symmetric,

$$
\boxed{
P(X\le0)=\frac12
}
$$

---

## (b) \(P(X>1)\)

$$
P(X>1)=1-\Phi(1)
$$

where \(\Phi\) is the standard normal CDF.

Numerically,

$$
\boxed{
P(X>1)\approx0.1587
}
$$

---

## (c) \(P(Y\in(-\infty,0]\times\mathbb R)\)

This event means

$$
T\le0.
$$

Thus,

$$
\boxed{
P(Y\in(-\infty,0]\times\mathbb R)=\frac12
}
$$

---

## (d) \(P(Y\in[-1,1]\times[-2,2])\)

Since \(T\) and \(P\) are independent,

$$
P(|T|\le1,\ |P|\le2)
=
P(|T|\le1)P(|P|\le2).
$$

Now,

$$
P(|T|\le1)=2\Phi(1)-1\approx0.6827
$$

and since \(P\sim N(0,4)\), we have \(P/2\sim N(0,1)\), so

$$
P(|P|\le2)=P(|Z|\le1)\approx0.6827.
$$

Therefore,

$$
\boxed{
P(Y\in[-1,1]\times[-2,2])
\approx
0.6827^2
\approx0.466
}
$$

---

# 8. Conditional Distribution of \(X\) Given \(Y=(t,p)\)

If the full state \(Y=(t,p)\) is known, then the value of \(X\) is completely determined:

$$
X=t.
$$

Hence,

$$
\boxed{
P(X=t\mid Y=(t,p))=1
}
$$

So the conditional distribution is degenerate at \(t\).

---

# 9. Conditional Distribution of \(Y\) Given \(X=t\)

Knowing \(X=t\) means the temperature is fixed.

The pressure remains random.

Since \(T\) and \(P\) are independent,

the conditional distribution is

$$
Y\mid X=t
=
(t,P),
$$

where

$$
P\sim N(0,4).
$$

Thus,

$$
\boxed{
Y\mid X=t
=
(t,P),
\quad
P\sim N(0,4)
}
$$

---

# 10. Interpretation

Conditioning on \(X\):

- reveals only the temperature deviation.

Conditioning on \(Y\):

- reveals both temperature and pressure deviations exactly.

Thus conditioning on \(Y\) gives strictly more information.

---

# 11. Show that \(X=\pi_1\circ Y\)

Define

$$
\pi_1(t,p)=t.
$$

Then,

$$
(\pi_1\circ Y)(t,p)
=
\pi_1(Y(t,p)).
$$

Since

$$
Y(t,p)=(t,p),
$$

we get

$$
(\pi_1\circ Y)(t,p)
=
\pi_1(t,p)
=
t.
$$

But

$$
X(t,p)=t.
$$

Therefore,

$$
\boxed{
X=\pi_1\circ Y
}
$$

Since \(\pi_1\) is measurable and \(Y\) is measurable,

\(X\) is a measurable function of \(Y\).

---



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# ---------------------------------------------------
# 1. Joint Density of Y = (T, P)
# ---------------------------------------------------

# Create grid values
t = np.linspace(-4, 4, 300)
p = np.linspace(-8, 8, 300)

T, P = np.meshgrid(t, p)

# Joint density function
f = (1/(4*np.pi)) * np.exp(-(T**2)/2 - (P**2)/8)

# Contour plot
plt.figure(figsize=(7,5))

contour = plt.contourf(T, P, f, levels=30)

plt.colorbar(contour)

plt.xlabel('Temperature deviation T')
plt.ylabel('Pressure deviation P')

plt.title('Joint Density of Y = (T,P)')

plt.show()


# ---------------------------------------------------
# 2. Marginal Density of X
# ---------------------------------------------------

x = np.linspace(-4, 4, 400)

# Standard normal density
fx = norm.pdf(x, loc=0, scale=1)

plt.figure(figsize=(7,5))

plt.plot(x, fx, linewidth=2)

plt.xlabel('x')
plt.ylabel('Density')

plt.title('Marginal Density of X ~ N(0,1)')

plt.grid(True)

plt.show()


# ---------------------------------------------------
# 3. Illustration of sigma(X) being coarser than sigma(Y)
# ---------------------------------------------------

# Generate random samples
np.random.seed(0)

T_samples = np.random.normal(0, 1, 400)
P_samples = np.random.normal(0, 2, 400)

plt.figure(figsize=(7,5))

# Scatter plot of full information Y
plt.scatter(T_samples, P_samples, alpha=0.5)

# Vertical regions illustrate sigma(X)
# X only sees temperature coordinate
plt.axvspan(-4, 0, alpha=0.15)
plt.axvspan(0, 4, alpha=0.15)

plt.xlabel('Temperature deviation T')
plt.ylabel('Pressure deviation P')

plt.title('sigma(X) is coarser than sigma(Y)')

plt.text(-3.5, 6, 'X only separates\nvertical regions')

plt.text(1.0, -6, 'Y distinguishes\nindividual points')

plt.show()
```
