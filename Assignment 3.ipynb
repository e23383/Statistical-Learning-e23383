# Q1) Prediction of Fraudulent Transactions: Detailed Solutions

1. Probability Space $(\Omega, \mathfrak{F}, P)$


The sample space is given as $\Omega = \{\omega_1, \omega_2, \omega_3, \omega_4\}$.


The $\sigma$-algebra $\mathfrak{F}$ is the power set $\mathfrak{P}(\Omega)$, which contains all $2^4 = 16$ possible subsets of $\Omega$:



$$\mathfrak{F} = \Big\{ \emptyset, \{\omega_1\}, \{\omega_2\}, \{\omega_3\}, \{\omega_4\}, \{\omega_1, \omega_2\}, \{\omega_1, \omega_3\}, \{\omega_1, \omega_4\}, \{\omega_2, \omega_3\}, \{\omega_2, \omega_4\}, \{\omega_3, \omega_4\}, \{\omega_1, \omega_2, \omega_3\}, \{\omega_1, \omega_2, \omega_4\}, \{\omega_1, \omega_3, \omega_4\}, \{\omega_2, \omega_3, \omega_4\}, \Omega \Big\}$$



The probability measure $P$ is fully defined by assigning probabilities to the singletons:


$P(\{\omega_1\}) = 0.50$


$P(\{\omega_2\}) = 0.10$


$P(\{\omega_3\}) = 0.20$


$P(\{\omega_4\}) = 0.20$


---


2. $\sigma$-algebra Generated by $Y$, namely $\sigma(Y)$


The random variable $Y$ maps the sample space to $\mathbb{R}$ as follows:


$Y^{-1}(\{0\}) = \{\omega_1, \omega_2\}$


$Y^{-1}(\{1\}) = \{\omega_3, \omega_4\}$



The $\sigma$-algebra $\sigma(Y)$ is the smallest $\sigma$-algebra containing these pre-images:



$$\sigma(Y) = \Big\{ \emptyset, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}, \Omega \Big\}$$


---



3. Measurability of $E[X \mid Y]$


By definition, the conditional expectation $E[X \mid Y]$ is the projection of $X$ onto the sub-$\sigma$-algebra $\sigma(Y)$. Therefore, $E[X \mid Y]$ must be $\sigma(Y)$-measurable so that it only relies on the information provided by the customer's risk class $Y$.



For a random variable to be measurable with respect to a specific $\sigma$-algebra, it must be constant over the "atoms" (the smallest non-empty disjoint building blocks) that generate that $\sigma$-algebra. Since the atoms of $\sigma(Y)$ are precisely $\{\omega_1, \omega_2\}$ and $\{\omega_3, \omega_4\}$, $E[X \mid Y]$ cannot distinguish between individual outcomes within these sets and must remain constant across them.


---


4. Compute $E[X \mid Y]$ Explicitly



Since $E[X \mid Y]$ is constant on the atoms of $\sigma(Y)$, we calculate its value for each group by taking the weighted average of $X$ over those outcomes:


On the set $\{\omega_1, \omega_2\}$ (where $Y=0$):



$$E[X \mid Y](\omega_1) = E[X \mid Y](\omega_2) = \frac{X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\})}{P(\{\omega_1, \omega_2\})} = \frac{0(0.50) + 1(0.10)}{0.50 + 0.10} = \frac{0.10}{0.60} = \frac{1}{6}$$



On the set $\{\omega_3, \omega_4\}$ (where $Y=1$):


$$E[X \mid Y](\omega_3) = E[X \mid Y](\omega_4) = \frac{X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\})}{P(\{\omega_3, \omega_4\})} = \frac{0(0.20) + 1(0.20)}{0.20 + 0.20} = \frac{0.20}{0.40} = \frac{1}{2}$$




Combining these using indicator functions yields:$$E[X \mid Y] = \frac{1}{6}\mathbb{1}_{\{Y=0\}} + \frac{1}{2}\mathbb{1}_{\{Y=1\}}$$

5. Verify the Defining Property of Conditional Expectation


We must verify that $\int_A E[X \mid Y] \, dP = \int_A X \, dP$ for all $A \in \sigma(Y)$.



Case 1: $A = \emptyset$
Both sides are trivially $0$.




Case 2: $A = \{\omega_1, \omega_2\}$



$$\int_A E[X \mid Y] \, dP = \frac{1}{6}P(\{\omega_1\}) + \frac{1}{6}P(\{\omega_2\}) = \frac{1}{6}(0.50 + 0.10) = 0.10$$



$$\int_A X \, dP = X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\}) = 0(0.50) + 1(0.10) = 0.10$$



Case 3: $A = \{\omega_3, \omega_4\}$



$$\int_A E[X \mid Y] \, dP = \frac{1}{2}P(\{\omega_3\}) + \frac{1}{2}P(\{\omega_4\}) = \frac{1}{2}(0.20 + 0.20) = 0.20$$



$$\int_A X \, dP = X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\}) = 0(0.20) + 1(0.20) = 0.20$$




Case 4: $A = \Omega$

$$\int_\Omega E[X \mid Y] \, dP = 0.10 + 0.20 = 0.30$$



$$\int_\Omega X \, dP = 0.10 + 0.20 = 0.30$$


The property holds true for every set in $\sigma(Y)$.

6. Conditional PMF $p_{X \mid Y}(x \mid y)$


The conditional probability mass function is given by $p_{X \mid Y}(x \mid y) = P(X=x \mid Y=y)$.




For $y = 0$:


$p_{X \mid Y}(0 \mid 0) = P(X=0 \mid Y=0) = \frac{P(\{\omega_1\})}{P(\{\omega_1, \omega_2\})} = \frac{0.50}{0.60} = \frac{5}{6}$



$p_{X \mid Y}(1 \mid 0) = P(X=1 \mid Y=0) = \frac{P(\{\omega_2\})}{P(\{\omega_1, \omega_2\})} = \frac{0.10}{0.60} = \frac{1}{6}$



For $y = 1$:



$p_{X \mid Y}(0 \mid 1) = P(X=0 \mid Y=1) = \frac{P(\{\omega_3\})}{P(\{\omega_3, \omega_4\})} = \frac{0.20}{0.40} = \frac{1}{2}$




$p_{X \mid Y}(1 \mid 1) = P(X=1 \mid Y=1) = \frac{P(\{\omega_4\})}{P(\{\omega_3, \omega_4\})} = \frac{0.20}{0.40} = \frac{1}{2}$

7. Verify Identity $\mathbb{E}[X \mid Y = y] = \sum_{x \in \{0,1\}} x \cdot p_{X \mid Y}(x \mid y)$



For $y = 0$:



$$\sum_{x \in \{0,1\}} x \cdot p_{X \mid Y}(x \mid 0) = 0 \cdot \frac{5}{6} + 1 \cdot \frac{1}{6} = \frac{1}{6}$$



This matches the value of $E[X \mid Y]$ when $Y=0$ found in Question 4.



For $y = 1$:

This matches the value of $E[X \mid Y]$ when $Y=1$ found in Question 4.

8. Calculate the Mean Square Error (MSE)


The Mean Square Error when predicting $X$ using the estimator $\hat{X} = E[X \mid Y]$ is equivalent to the expected conditional variance, $E[\text{Var}(X \mid Y)]$.


First, let's look at the errors squared ($(X - E[X \mid Y])^2$) for each outcome:


$\omega_1$: $(0 - \frac{1}{6})^2 = \frac{1}{36}$



$\omega_2$: $(1 - \frac{1}{6})^2 = \frac{25}{36}$


$\omega_3$: $(0 - \frac{1}{2})^2 = \frac{1}{4}$


$\omega_4$: $(1 - \frac{1}{2})^2 = \frac{1}{4}$


Now, we calculate the expected value of these squared errors:


$$\text{MSE} = \sum_{i=1}^4 (X(\omega_i) - E[X \mid Y](\omega_i))^2 \cdot P(\{\omega_i\})$$




$$\text{MSE} = \left(\frac{1}{36} \times 0.50\right) + \left(\frac{25}{36} \times 0.10\right) + \left(\frac{1}{4} \times 0.20\right) + \left(\frac{1}{4} \times 0.20\right)$$




$$\text{MSE} = \frac{0.5}{36} + \frac{2.5}{36} + 0.05 + 0.05 = \frac{3}{36} + 0.10 = \frac{1}{12} + \frac{1}{10} = \frac{11}{60} \approx 0.1833$$




---


9. Change in MSE Using Global Average & Relationship to Law of Total Variance


If the bank ignores $Y$ and uses the constant global baseline estimate $E[X] = 0.30$, the new error is measured by the total variance $\text{Var}(X)$.



Since $X$ is a Bernoulli random variable with $p = 0.30$:



$$\text{MSE}_{\text{global}} = \text{Var}(X) = p(1-p) = 0.30 \times 0.70 = 0.21$$


Relationship to the Law of Total Variance:

The Law of Total Variance states:


$$\text{Var}(X) = E[\text{Var}(X \mid Y)] + \text{Var}(E[X \mid Y])$$


Where:


$E[\text{Var}(X \mid Y)] = \text{MSE with } Y = \frac{11}{60} \approx 0.1833$



$\text{Var}(E[X \mid Y])$ represents the variance explained by knowing the risk class $Y$.



Let's compute $\text{Var}(E[X \mid Y])$ to verify:



$$E[E[X \mid Y]] = E[X] = 0.30$$



$$\text{Var}(E[X \mid Y]) = \left(\frac{1}{6} - 0.3\right)^2(0.60) + \left(\frac{1}{2} - 0.3\right)^2(0.40)$$


$$\text{Var}(E[X \mid Y]) = \left(-\frac{2}{15}\right)^2(0.60) + (0.2)^2(0.40) = \left(\frac{4}{225} \times \frac{3}{5}\right) + (0.04 \times 0.4)$$



$$\text{Var}(E[X \mid Y]) = \frac{4}{375} + 0.016 = 0.01067 + 0.016 = 0.02667 = \frac{4}{150} = \frac{1}{37.5}$$



Checking the law:



$$\text{Var}(X) = 0.18333 + 0.02667 = 0.21$$


Conclusion: Ignoring the customer risk class increases the MSE from $0.1833$ to $0.21$. The difference ($0.02667$) is exactly the variance reduction achieved by utilizing the descriptive power of $Y$.

10. (a)  Entropy $H(X)$



The indicator variable $X$ takes values in $\{0, 1\}$ with probabilities $P(X=1) = 0.30$ and $P(X=0) = 0.70$.


The information entropy $H(X)$ (using base 2 logarithms) is calculated as:


$$H(X) = - \sum_{x \in \{0,1\}} p(x) \log_2 p(x)$$



$$H(X) = - \left( 0.70 \log_2(0.70) + 0.30 \log_2(0.30) \right)$$



Using the change of base rule ($\log_2(n) = \frac{\ln(n)}{\ln(2)}$):


$\log_2(0.70) \approx -0.5146$


$\log_2(0.30) \approx -1.7370$



$$H(X) \approx - \left( 0.70(-0.5146) + 0.30(-1.7370) \right) = - ( -0.3602 - 0.5211 ) = 0.8813 \text{ bits}$$



---




b) Compute the Conditional Entropy $H(X \mid Y)$


Conditional entropy measures the average uncertainty remaining about the fraud status $X$ after the bank observes the customer's risk class $Y$. The formula is:



$$H(X \mid Y) = \sum_{y \in \{0,1\}} P(Y=y) H(X \mid Y=y)$$


First, we calculate the entropy of $X$ given each specific value of $Y$ using the conditional probabilities found in Question 6:



For $y = 0$ (Low-risk): $P(X=0 \mid Y=0) = \frac{5}{6}$ and $P(X=1 \mid Y=0) = \frac{1}{6}$




$$H(X \mid Y=0) = -\left( \frac{5}{6}\log_2\left(\frac{5}{6}\right) + \frac{1}{6}\log_2\left(\frac{1}{6}\right) \right) \approx -(-0.2183 - 0.4308) = 0.6500\text{ bits}$$


For $y = 1$ (High-risk): $P(X=0 \mid Y=1) = \frac{1}{2}$ and $P(X=1 \mid Y=1) = \frac{1}{2}$



$$H(X \mid Y=1) = -\left( \frac{1}{2}\log_2\left(\frac{1}{2}\right) + \frac{1}{2}\log_2\left(\frac{1}{2}\right) \right) = -(-0.5 - 0.5) = 1.0000\text{ bits}$$


Now, weight these by the marginal probabilities of $Y$ ($P(Y=0) = 0.60$ and $P(Y=1) = 0.40$):


$$H(X \mid Y) = (0.60 \times 0.6500) + (0.40 \times 1.0000)$$



$$H(X \mid Y) = 0.3900 + 0.4000 = 0.7900\text{ bits}$$


---


c) Calculate the Mutual Information $I(X; Y)$


Mutual Information represents the amount of information shared between $X$ and $Y$, which corresponds to the reduction in uncertainty about fraud when the risk class is known:



$$I(X; Y) = H(X) - H(X \mid Y)$$



Using our calculated values:



$$I(X; Y) = 0.8813 - 0.7900 = 0.0913\text{ bits}$$




---




d) Percentage of Total Uncertainty Removed



To find the percentage of total uncertainty about fraud that is eliminated by knowing the customer's risk class, we calculate the ratio of the shared information to the initial uncertainty:


$$\text{Percentage Removed} = \frac{I(X; Y)}{H(X)} \times 100\%$$



$$\text{Percentage Removed} = \frac{0.0913}{0.8813} \times 100\% \approx 10.36\%$$



Conclusion: Knowing whether a customer belongs to the low-risk or high-risk category eliminates roughly 10.36% of the bank's initial uncertainty regarding whether a transaction is fraudulent.

# Q2) Conditional Density of the Multivariate Gaussian


To derive the conditional density $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, we use the definition of conditional probability density:


$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$$



Instead of manually expanding the massive joint matrix fraction directly, the most elegant way to solve this is by using a linear transformation to decouple the variables $\mathbf{X}$ and $\mathbf{Y}$.



Step 1: Introduce a Decoupling Transformation


We want to transform $\mathbf{X}$ into a new random vector $\mathbf{X}^*$ that is completely uncorrelated (and thus independent) from $\mathbf{Y}$. Let's define:



$$\mathbf{X}^* = \mathbf{X} - \mathbf{A}\mathbf{Y}$$



We choose the matrix $\mathbf{A}$ such that the cross-covariance $\text{Cov}(\mathbf{X}^*, \mathbf{Y}) = \mathbf{0}$:



$$\text{Cov}(\mathbf{X} - \mathbf{A}\mathbf{Y}, \mathbf{Y}) = \text{Cov}(\mathbf{X}, \mathbf{Y}) - \mathbf{A}\text{Var}(\mathbf{Y}) = \mathbf{0}$$




$$\mathbf{\Sigma}_{XY} - \mathbf{A}\mathbf{\Sigma}_{YY} = \mathbf{0} \implies \mathbf{A} = \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}$$



Thus, our transformed variable is:



$$\mathbf{X}^* = \mathbf{X} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{Y}$$


Step 2: Find the Distribution of the Transformed System


Since $\mathbf{X}^*$ and $\mathbf{Y}$ are linear combinations of the joint multivariate normal vector $\mathbf{Z}$, they are also jointly multivariate normal. Because we forced their cross-covariance to be zero, $\mathbf{X}^*$ and $\mathbf{Y}$ are statistically independent.


Let's find the mean and variance of $\mathbf{X}^*$:


1. Mean ($\mathbf{\mu}^*$):


$$\mathbf{\mu}_{X^*} = E[\mathbf{X}^*] = E[\mathbf{X} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{Y}] = \mathbf{\mu}_X - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\mu}_Y$$


2. Covariance ($\mathbf{\Sigma}_{X^*X^*}$):


$$\mathbf{\Sigma}_{X^*X^*}= \text{Var}(\mathbf{X} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{Y})$$



$$\mathbf{\Sigma}_{X^*X^*} = \text{Var}(\mathbf{X}) - \text{Cov}(\mathbf{X}, \mathbf{Y})\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\text{Cov}(\mathbf{Y}, \mathbf{X}) + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\text{Var}(\mathbf{Y})\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX}$$


$$\mathbf{\Sigma}_{X^*X^*} = \mathbf{\Sigma}_{XX} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX} + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX}$$




$$\mathbf{\Sigma}_{X^*X^*} = \mathbf{\Sigma}_{XX} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX}$$



Notice that this matches the definition of the Schur complement $\mathbf{\Sigma}_{X|Y}$ given in the problem statement.


Step 3: Evaluate the Conditional Distribution


Now we look at the conditional expectation and distribution. Because $\mathbf{X}^*$ and $\mathbf{Y}$ are independent, the conditional distribution of $\mathbf{X}^*$ given $\mathbf{Y} = \mathbf{y}$ is simply its marginal distribution:



$$\mathbf{X}^* \mid \mathbf{Y}=\mathbf{y} \sim \mathcal{N}(\mathbf{\mu}_{X^*}, \mathbf{\Sigma}_{X^*X^*})$$



Since $\mathbf{X}^* = \mathbf{X} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{Y}$, we can isolate $\mathbf{X}$ when given $\mathbf{Y} = \mathbf{y}$:



$$\mathbf{X} \mid \mathbf{Y}=\mathbf{y} = \mathbf{X}^* + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{y}$$



Adding a constant vector to a Gaussian distribution only shifts its mean and leaves the covariance untouched. Therefore:




Conditional Mean ($\mathbf{\mu}^*$):




$$\mathbf{\mu}^* = E[\mathbf{X} \mid \mathbf{Y}=\mathbf{y}] = E[\mathbf{X}^*] + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{y}$$



$$\mathbf{\mu}^* = \mathbf{\mu}_X - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\mu}_Y + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{y}$$


$$\mathbf{\mu}^* = \mathbf{\mu}_X + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}(\mathbf{y} - \mathbf{\mu}_Y)$$



Conditional Covariance ($\mathbf{\Sigma}_{X|Y}$):




$$\mathbf{\Sigma}_{X|Y} = \text{Var}(\mathbf{X} \mid \mathbf{Y}=\mathbf{y}) = \text{Var}(\mathbf{X}^*) = \mathbf{\Sigma}_{XX} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX}$$



Conclusion: Writing down the PDF


Since $\mathbf{X} \mid \mathbf{Y}=\mathbf{y}$ is a multivariate normal distribution of dimension $n$ with mean $\mathbf{\mu}^*$ and covariance matrix $\mathbf{\Sigma}_{X|Y}$, substituting these parameters directly into the standard multivariate normal density formula yields:




$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n|\mathbf{\Sigma}_{X|Y}|}}\exp\left( -\frac{1}{2}(\mathbf{x} - \mathbf{\mu}^*)^T\mathbf{\Sigma}_{X|Y}^{-1}(\mathbf{x} - \mathbf{\mu}^*) \right)$$




where:



$\mathbf{\Sigma}_{X|Y} = \mathbf{\Sigma}_{XX} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX}$




$\mathbf{\mu}^* = \mathbf{\mu}_x + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}(\mathbf{y} - \mathbf{\mu}_y)$




This completes the proof.

# Q3) Continuous Fault Monitoring of a Spindle Motor



Since $X$ and $Q$ are independent normal random variables, any linear combination of them is also normally distributed. Because $Y = X + Q$, the vector $(X, Y)$ follows a bivariate normal distribution.



Show that $Y \sim \mathcal{N}(0, 5)$


Mean:


$$E[Y] = E[X + Q] = E[X] + E[Q] = 0 + 0 = 0$$



Variance: Since $X$ and $Q$ are independent, their variances add directly:


$$\text{Var}(Y) = \text{Var}(X + Q) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$$



Thus, $Y \sim \mathcal{N}(0, 5)$.


Find the covariance $\text{Cov}(X, Y)$


$$\text{Cov}(X, Y) = \text{Cov}(X, X + Q) = \text{Cov}(X, X) + \text{Cov}(X, Q)$$



Since $\text{Cov}(X, X) = \text{Var}(X) = 1$ and $X, Q$ are independent ($\text{Cov}(X, Q) = 0$):





$$\text{Cov}(X, Y) = 1 + 0 = 1$$



---


2. Information Geometry & Measurability


For $X$ to be $\sigma(Y)$-measurable, its value would have to be completely and uniquely determined by observing $Y$. This is not the case because multiple distinct pairs of $(X, Q)$ can produce the exact same sum $Y$.



Practical Interpretation:




If the power meter displays $Y = 5$, that value is a shared outcome of both thermal changes and mechanical resistance. An engineer cannot tell for certain if the motor is overheating ($X = 4, Q = 1$) or if the spindle is just cutting an exceptionally hard material ($X = 0, Q = 5$). Because $Y$ confounds these two independent hidden factors, observing $Y$ leaves an irreducible degree of uncertainty regarding the exact value of $X$.




---


3. Conditional Expectation (The Best Estimate)


Using the linear regression property for joint normals:



$$E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$$



Substituting our known values ($\mu_X = 0$, $\mu_Y = 0$, $\text{Cov}(X,Y) = 1$, $\text{Var}(Y) = 5$):



$$E[X \mid Y = y] = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y$$


Expressed as a "Signal-to-Total-Variance" Ratio:


We can write the coefficient by substituting the fundamental variance components:



$$\frac{\text{Cov}(X,Y)}{\text{Var}(Y)} = \frac{\text{Var}(X)}{\text{Var}(X) + \text{Var}(Q)} = \frac{\sigma^2_{\text{signal}}}{\sigma^2_{\text{signal}} + \sigma^2_{\text{noise}}}$$


This shows our optimal estimator weights the observation $y$ based on how clean the signal is relative to the total noise background.




---


4. Verification of the Residual


Let $\hat{X} = E[X \mid Y] = \frac{1}{5}Y$. The residual error is defined as $Z = X - \hat{X} = X - \frac{1}{5}Y$.


Prove that $E[Z] = 0$


$$E[Z] = E\left[X - \frac{1}{5}Y\right] = E[X] - \frac{1}{5}E[Y] = 0 - \frac{1}{5}(0) = 0$$


Verify that error $Z$ is uncorrelated with observation $Y$ (Orthogonality Principle)


$$\text{Cov}(Z, Y) = \text{Cov}\left(X - \frac{1}{5}Y, Y\right) = \text{Cov}(X, Y) - \frac{1}{5}\text{Cov}(Y, Y)$$



$$\text{Cov}(Z, Y) = \text{Cov}(X, Y) - \frac{1}{5}\text{Var}(Y)$$


Substituting $\text{Cov}(X,Y) = 1$ and $\text{Var}(Y) = 5$:



$$\text{Cov}(Z, Y) = 1 - \frac{1}{5}(5) = 1 - 1 = 0$$



Since the covariance is $0$, the estimation error $Z$ is completely orthogonal (uncorrelated) to our observation space $Y$.



---




5. Information Theory: Differential Entropy


The differential entropy of a univariate normal distribution $\mathcal{N}(\mu, \sigma^2)$ is given by:



$$h = \frac{1}{2}\ln(2\pi e \sigma^2)$$


Compute $h(X)$



Since $\text{Var}(X) = 1$:



$$h(X) = \frac{1}{2}\ln(2\pi e (1)) = \frac{1}{2}\ln(2\pi e)$$


Compute $h(X \mid Y)$



First, find the conditional variance $\sigma^2_{X|Y}$:


$$\sigma^2_{X|Y} = \sigma^2_X - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1^2}{5} = 1 - 0.20 = 0.80$$



Using this conditional variance to compute the conditional differential entropy:



$$h(X \mid Y) = \frac{1}{2}\ln(2\pi e \sigma^2_{X|Y}) = \frac{1}{2}\ln(2\pi e (0.80))$$



---




6. Mutual Information $I(X; Y)$



$$I(X; Y) = h(X) - h(X \mid Y)$$



$$I(X; Y) = \frac{1}{2}\ln(2\pi e) - \frac{1}{2}\ln(2\pi e \cdot 0.80) = \frac{1}{2}\ln\left(\frac{2\pi e}{2\pi e \cdot 0.80}\right) = \frac{1}{2}\ln\left(\frac{1}{0.80}\right) = \frac{1}{2}\ln(1.25) \approx 0.1116\text{ nats}$$



Percentage of uncertainty removed:


In information theory, the relative reduction in continuous variance uncertainty can be assessed directly by looking at the variance reduction percentage:




$$\text{Uncertainty Reduction} = \frac{\sigma^2_X - \sigma^2_{X|Y}}{\sigma^2_X} \times 100\% = \frac{1 - 0.80}{1} \times 100\% = 20\%$$



How would this change if $\text{Var}(Q) = 100$?


If the mechanical noise variance increases to $100$:


$\text{Var}(Y) = 1 + 100 = 101$



$\text{Cov}(X,Y) = 1$



New conditional variance: $\sigma^2_{X|Y} = 1 - \frac{1^2}{101} = \frac{100}{101} \approx 0.9901$




The new variance reduction drops significantly:



$$\text{New Uncertainty Reduction} = \frac{1 - 0.9901}{1} \times 100\% \approx 0.99\%$$





Observing the power meter now reveals almost no useful information because the signal is completely drowned out by the huge mechanical noise.

7. Diagnostic Interpretation




The factory manager's intuition that $\hat{X} = Y$ is incorrect because it completely ignores the noise component $Q$.


If we set our estimate to $\hat{X} = Y$, the expected value of our estimate given a specific true value of $X$ is:




$$E[Y \mid X] = E[X + Q \mid X] = X + 0 = X$$



While it tracks $X$ on average across all possible cuts, its conditional variance as an estimator is highly problematic because $\text{Var}(Y \mid X) = \text{Var}(Q) = 4$. This means any single reading is wildly unstable due to mechanical load changes.



Instead, the true MMSE conditional expectation expression is:



$$E[X \mid Y] = \frac{1}{5}Y$$




This expression "shrinks" the direct raw observation $Y$ toward the prior baseline mean (zero). Because the power reading is heavily corrupted by mechanical noise ($\text{Var}(Q)=4$) relative to the actual temperature variations ($\text{Var}(X)=1$), any extreme spike in $Y$ is statistically much more likely to be caused by a temporary spike in cutting resistance rather than a sudden thermal emergency. The shrinkage factor $\frac{1}{5}$ acts as a mathematical stabilizer, pulling back rash observations to prevent false alarms.
